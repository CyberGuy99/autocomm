```python
from autocomm_v1.experiment import CircuitGen
from annotating_circuit import get_circuit_input as gci

BV = CircuitGen.BV
QFT = CircuitGen.QFT
RCA = CircuitGen.RCA
```

```python
b_gl,b_m = BV(100, 10)
r_gl,r_m = RCA(100, 10)
q_gl,q_m = QFT(100, 10)
```

```python
gci(b_gl, b_m)
```

```python
gci(r_gl, r_m)
```

```python
gci(q_gl, q_m)
```

```python
qpu_qubit_counts = {
        'QPU1': {'computing': 50, 'communication': 1},  # QPU1 has 4 computing qubits, 2 communication qubits
        'QPU2': {'computing': 50, 'communication': 1},  # QPU2 has 3 computing qubits, 1 communication qubit
    }
```

```python
from annotating_circuit import simulate
b_autocommed = gci(b_gl, b_m)
q_autocommed = gci(q_gl, q_m)
r_autocommed = gci(r_gl, r_m)


circuits = [b_autocommed, q_autocommed, r_autocommed]
simulate(circuits, qpu_qubit_counts)
```

```python
from qubit_partition import pymetis_partition as partition

b_autocommed = gci(b_gl, partition(num_nodes=10, gate_list=b_gl))
q_autocommed = gci(q_gl, partition(num_nodes=10, gate_list=q_gl))
r_autocommed = gci(r_gl, partition(num_nodes=10, gate_list=r_gl))


circuits = [b_autocommed, q_autocommed, r_autocommed]
simulate(circuits, qpu_qubit_counts)
```
